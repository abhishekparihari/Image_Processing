```python
import pytesseract
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#initialising Tesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract"
```

```python
#Reading image from the folder using openCv which will be in BGR format
input_image = cv2.imread('../Image Processing/Images/GIT.png')
```

```python
#Showing and plotting the BGR image 
plt.imshow(input_image)
plt.title('Input Text Image in BGR mode')
plt.show()
```

```python
#Reading image from the folder using pillow lib 
image_pil= Image.open('../Image Processing/Images/GIT.png')
plt.imshow(image_pil)
plt.title('Input Text Image using Pillow lib')
plt.show()
```

```python
#converting image to RGB. By default in openCv it's present as BGR.
txt_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
```

```python
#Showing and plotting the RGB image which human eyes basically see.
plt.imshow(txt_image)
plt.title('Input Text Image after RGB conversion')
plt.show()
```

```python
#extracting text present in provided image
text_present_in_image = pytesseract.image_to_string(txt_image)
print('*****::-- Text Extracted from provied Image --::*****')
print(text_present_in_image)
```
